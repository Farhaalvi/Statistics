# What is Central Limit Theorem?

The Central Limit Theorem (CLT) is a fundamental concept in statistics and probability theory. It states that when independent random variables are added, their sum tends to follow a normal distribution, regardless of the shape of the original variables' distributions.

In other words, if you have a large enough sample size and you calculate the mean of a sufficiently large number of independent and identically distributed random variables, **the distribution of those sample means will be approximately normal, regardless of the shape of the original population** from which the samples are drawn.

The Central Limit Theorem is significant because it allows us to make inferences and perform statistical tests based on the assumption of normality, even if the underlying population distribution is not normal. It forms the foundation for many statistical techniques, including hypothesis testing, confidence intervals, and regression analysis.

To summarize, the Central Limit Theorem states that the sample means from a large number of independent and identically distributed random variables will tend to be normally distributed, providing an important tool for statistical analysis.

## Uniform Distribution

This uniform distribution has a lower bound of 0 and an upder bound of 100.


```python
# Set the seed value

random.seed(42)

population_uniform = [random.uniform(0,100) for iter in range(1000000)]
```

```python
population_uniform_mean = np.mean(population_uniform)
population_uniform_sd = np.std(population_uniform)
```

```python
print(f"The population follows an uniform distribution between 0 and 100. The population list has {len(population_uniform)} elements. \n\nThe population mean is: {population_uniform_mean: .2f} \nThe population standard deviation is: {population_uniform_sd: .2f}")
```

### Visualizing the uniform distribution


```python
# Plot the histogram

plt.hist(population_uniform, bins = 100, 
         density = True, alpha = 0.7, color = 'blue')
plt.xlabel('X = x')
plt.ylabel('Frequency')
plt.title('Uniform Distribution')
plt.axvline(population_uniform_mean, color='red', 
            linestyle='dashed', linewidth=2, label= f"Mean = {population_uniform_mean:.2f}")
plt.legend(loc = "lower right")

# Adjust the spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.grid(True)
plt.show()
```

### Random Samples

#### Sample Size = 100


```python
# Function for Uniform  Sample mean

def sample_uniform_mean(population, sample_number, sample_size = 100):
    
    sample_mean = []
    
    for iter in range(sample_number):
        # Ramdomly choose sample_size elements from the population
        sample = random.sample(population, sample_size)
        mean = np.mean(sample)
        sample_mean.append(mean)
        
    return sample_mean
```

```python
# Functions for sample_mean histograms and bell-curve:


def plot_sample_mean_hist(sample_mean, bins = 'rice'):
    # Plot the histogram
    plt.hist(sample_mean, bins = bins, 
             density=True, alpha=0.7, 
             label=f'Histograms for {len(sample_mean)} Sample Means')
    
    # Fit a normal distribution to the sample means
    mu, std = norm.fit(sample_mean)

    # Generate x-axis values for the bell curve
    x = np.linspace(min(sample_mean), 
                max(sample_mean), 1000)
    # Calculate y-axis values for the bell curve
    y = norm.pdf(x, mu, std)
    
    # Plot the bell curve
    plt.plot(x, y, 'r-', label='Normal Distribution')

    plt.xlabel('Sample Means')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {len(sample_mean)} Sample Means')
    plt.legend()

    plt.grid(True)

    plt.show()
```

```python
# Number of samples = 20

sample_uniform_mean_20 = sample_uniform_mean(population_uniform, sample_number= 20, sample_size = 100 )
plot_sample_mean_hist(sample_uniform_mean_20)
```

## Logarithmic Distribution

```python
# Set the random seed
np.random.seed(42)

# Generate random numbers from logarithmic distribution
population_log = np.random.logseries(p=0.5, size=1000000)
```

```python
population_log_mean = np.mean(population_log)
population_log_sd = np.std(population_log)
```

```python
print(f"The population follows a logarithmic distribution. The population list has {len(population_log)} elements. \n\nThe population mean is: {population_log_mean: .2f} \nThe population standard deviation is: {population_log_sd: .2f}")
```

### Visualizing the logarithmic distribution

```python
# Generate x-values for the PDF curve
x = np.arange(1, np.max(population_log) + 1)

# Calculate the PDF values for the logarithmic distribution
pdf = logser.pmf(x, p=0.5)

# Plot the logarithmic curve
plt.plot(x, pdf, 'b-', label='Logarithmic Distribution')

# Plot the mean as a vertical line
plt.axvline(x=population_log_mean, color='green', 
            linestyle='--', label= f'Mean: {population_log_mean: .2f}')


plt.xlabel('X = x')
plt.ylabel('Probability Density')
plt.title('Logarithmic Distribution Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Function for logarithmic Sample mean
def sample_log_mean(population, sample_number, sample_size = 100):
    sample_mean = []
    
    for iter in range(sample_number):
        # Ramdomly choose sample_size elements from the population
        sample = np.random.choice(population, 
                                  size=sample_size, replace=False)
        mean = np.mean(sample)
        sample_mean.append(mean)
    return sample_mean
    
```
